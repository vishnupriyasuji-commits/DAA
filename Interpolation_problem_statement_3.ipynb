```python
import time
import random

def interpolation_search(arr, key):
    """
    Performs interpolation search on a sorted array.
    Returns: (found: bool, index: int, probes: int, exec_time: float)
    """
    start_time = time.perf_counter()
    low = 0
    high = len(arr) - 1
    probes = 0
    found = False
    index = -1

    # The key must be within the bounds of the array
    while low <= high and arr[low] <= key <= arr[high]:
        probes += 1

        # Corner case: if all elements are identical
        if arr[high] == arr[low]:
            if arr[low] == key:
                found = True
                index = low
            break

        # Interpolation formula
        pos = low + int(((high - low) / (arr[high] - arr[low])) * (key - arr[low]))

        # Safeguard for index bounds due to floating point rounding
        if pos < low or pos > high:
            break

        if arr[pos] == key:
            found = True
            index = pos
            break
        elif arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    end_time = time.perf_counter()
    exec_time = (end_time - start_time) * 1000  # Convert to milliseconds

    return found, index, probes, exec_time

def generate_datasets(size, choice):
    """Generates either a uniform or non-uniform sorted dataset."""
    if choice == 1:
        # Uniform: Constant step increments
        step = random.randint(2, 10)
        start = random.randint(1, 100)
        return [start + i * step for i in range(size)]
    elif choice == 2:
        # Non-Uniform: Exponential or heavily skewed growth
        arr = []
        val = random.randint(1, 5)
        for _ in range(size):
            arr.append(val)
            val += random.randint(1, 5) if random.random() > 0.1 else random.randint(100, 500)
        return arr
    return []

def main():
    print("=== Interpolation Search Performance Analyzer ===")
    print("1. Uniformly Distributed Dataset")
    print("2. Non-Uniformly Distributed Dataset")

    try:
        dist_choice = int(input("Choose distribution type (1 or 2): "))
        if dist_choice not in [1, 2]:
            print("Invalid choice!")
            return

        mode_choice = input("Generate data randomly? (y/n): ").strip().lower()

        if mode_choice == 'y':
            size = int(input("Enter dataset size (e.g., 1000): "))
            dataset = generate_datasets(size, dist_choice)
        else:
            user_input = input("Enter sorted elements separated by spaces: ")
            dataset = sorted([int(x) for x in user_input.split()])

        print(f"\nGenerated/Sorted Dataset (First 15 elements): {dataset[:15]}...")

        key = int(input("\nEnter the search key: "))

        # Perform Search
        found, index, probes, exec_time = interpolation_search(dataset, key)

        # Display Results
        print("\n--- Search Results ---")
        print(f"Key Found:      {found}")
        print(f"Index Position: {index if found else 'N/A'}")
        print(f"Number of Probes: {probes}")
        print(f"Execution Time:  {exec_time:.4f} ms")

    except ValueError:
        print("Please enter valid numerical inputs.")

if __name__ == "__main__":
    main()
```
